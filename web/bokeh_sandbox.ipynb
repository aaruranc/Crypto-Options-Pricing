```python
cwd = os.getcwd()
print(cwd)

df =  pd.read_csv('2-Day.csv')
print(df)
```

```python
d = 'Date'
n = '90%-Spot-Vol-Mean-B-S-Call'
date = df[d]
price = df[n]
print(date)
print(price)
print(type(date))
```

```python
d = len(price)
l = []
for i in range(0,d):
    l.append(i)
print(l)

```

```python
from bokeh.plotting import figure, output_file, show

output_file("line.html")

p = figure(plot_width=400, plot_height=400)
p.line(l, price, line_width=2)

show(p)
```
